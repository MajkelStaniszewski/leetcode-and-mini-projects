```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/lego_sets.csv")

theme = pd.read_csv("https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/parent_themes.csv")
```

```python
merged = df.merge(theme, left_on = 'parent_theme',right_on ='name')
merged.drop(columns ='name_y', inplace= True)
merged.head()
```

```python
# 1. What percentage of all licensed set ever released were Star Wars themed?
licensed = merged[merged['is_licensed']]
licensed = licensed.dropna(subset = ['set_num'])
star_wars = licensed[licensed['parent_theme'] == 'Star Wars']
the_force = round((star_wars.shape[0]/licensed.shape[0]) * 100,2)
print(f'Star Wars themed licenses percentage: {the_force} %')
#merged[merged['set_num'].isnull()].shape
```

```python
# 2. In which year was Star Wars not the most popular licensed theme (in terms of number of sets released that year)?
# Save your answer as an variable new_era in the form of an integer
licensed.sort_values('year', inplace = True)
licensed['count'] = 1
licensed.head()
xd = licensed.groupby(['year','parent_theme']).sum().reset_index()
xd.sort_values('count',ascending = False, inplace = True)
xd = xd.drop_duplicates(['year'])
xd.sort_values('year', inplace = True)
xd
# new_era = 2017
```

```python
# 3. Break down number of sets by year
clean_df = merged[~merged['set_num'].isnull()]
clean_df['count'] = 1
sets_per_year = clean_df.groupby(['year']).sum().reset_index()[['year','count']]
#sets_per_year.head()
for index,row in sets_per_year.iterrows():
  print(row['year'],row['count'])
```
